```python
full_data = [train, test]

# Некоторые собственные функции, которые я добавил в
# Дает длину имени
train['Name_length'] = train['Name'].apply(len)
test['Name_length'] = test['Name'].apply(len)
#train.head()
```

```python
# Функция, которая говорит, была ли у пассажира кабина на Титанике
train['Has_Cabin'] = train["Cabin"].apply(lambda x: 0 if type(x) == float else 1)
test['Has_Cabin'] = test["Cabin"].apply(lambda x: 0 if type(x) == float else 1)
train.head()
```

```python
# Особенности инженерных шагов, взятых из Сина
# Создайте новую функцию FamilySize как комбинацию SibSp и Parch
for dataset in full_data:
    dataset['FamilySize'] = dataset['SibSp'] + dataset['Parch'] + 1
#train.head()
#test.head()
```

```python
# Создать новую функцию IsAlone от FamilySize
for dataset in full_data:
    dataset['IsAlone'] = 0
    dataset.loc[dataset['FamilySize'] == 1, 'IsAlone'] = 1
train.head()
#test.head()
```

```python
# Удалить все NULL в столбце Embarked
for dataset in full_data:
    dataset['Embarked'] = dataset['Embarked'].fillna('S')
#train.head()
#test.head()
```

```python
# Удалите все NULLS в столбце «Тариф» и создайте новую функцию.
for dataset in full_data:
    dataset['Fare'] = dataset['Fare'].fillna(train['Fare'].median())
train['CategoricalFare'] = pd.qcut(train['Fare'], 4)
#train.head()
#test.head()
```

```python
# Создать новую функцию Категориальный возраст
for dataset in full_data:
    age_avg = dataset['Age'].mean()
    #print(age_avg)
    age_std = dataset['Age'].std()
    #print(age_std)
    age_null_count = dataset['Age'].isnull().sum()
    #print(age_null_count)
    age_null_random_list = np.random.randint(age_avg - age_std, age_avg + age_std, size=age_null_count)
    #print(np.random.randint(age_avg - age_std, age_avg + age_std, size=age_null_count))
    dataset['Age'][np.isnan(dataset['Age'])] = age_null_random_list
    dataset['Age'] = dataset['Age'].astype(int)
    #print(dataset['Age'].isnull().sum())
train['CategoricalAge'] = pd.cut(train['Age'], 5)

#train.head()
test.head()
```

```python
# Определить функцию для извлечения названий из имен пассажиров
def get_title(name):
    #print(name)
    title_search = re.search(' ([A-Za-z]+)\.', name)
    # Если заголовок существует, извлеките и верните его.
    if title_search:
        return title_search.group(1)
    return ""
```

```python
# Все не общие названия группируются в единую группировку "Редкие"
for dataset in full_data:
    dataset['Title'] = dataset['Title'].replace(['Lady', 'Countess','Capt', 'Col','Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')

    dataset['Title'] = dataset['Title'].replace('Mlle', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Ms', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Mme', 'Mrs')

train.head()
#test.head()
```

```python
for dataset in full_data:
    # Картирование пола
    dataset['Sex'] = dataset['Sex'].map( {'female': 0, 'male': 1} ).astype(int)
    
    # Названия карт
    title_mapping = {"Mr": 1, "Miss": 2, "Mrs": 3, "Master": 4, "Rare": 5}
    dataset['Title'] = dataset['Title'].map(title_mapping)
    dataset['Title'] = dataset['Title'].fillna(0)
    
    # Mapping путь начинал
    dataset['Embarked'] = dataset['Embarked'].map( {'S': 0, 'C': 1, 'Q': 2} ).astype(int)
    
    # Создание картографии
    dataset.loc[ dataset['Fare'] <= 7.91, 'Fare'] = 0
    dataset.loc[(dataset['Fare'] > 7.91) & (dataset['Fare'] <= 14.454), 'Fare'] = 1
    dataset.loc[(dataset['Fare'] > 14.454) & (dataset['Fare'] <= 31), 'Fare']   = 2
    dataset.loc[ dataset['Fare'] > 31, 'Fare'] = 3
    dataset['Fare'] = dataset['Fare'].astype(int)
    
    # Возраст картографирования
    dataset.loc[ dataset['Age'] <= 16, 'Age'] = 0
    dataset.loc[(dataset['Age'] > 16) & (dataset['Age'] <= 32), 'Age'] = 1
    dataset.loc[(dataset['Age'] > 32) & (dataset['Age'] <= 48), 'Age'] = 2
    dataset.loc[(dataset['Age'] > 48) & (dataset['Age'] <= 64), 'Age'] = 3
    dataset.loc[ dataset['Age'] > 64, 'Age'] = 4 ;

train.head()
#test.head()
```

```python
Помощники через классы Python
Здесь мы вызываем использование классов Python, чтобы сделать его более удобным для нас. Любой новичок в 
программировании обычно слышит, как используются классы в сочетании с объектно-ориентированным 
программированием (ООП). Короче говоря, класс помогает расширить некоторый код / программу для создания 
объектов (переменные для просмотра старой школы), а также для реализации функций и методов, специфичных 
для этого класса.

В следующем разделе кода мы, по сути, пишем класс SklearnHelper, который позволяет расширять встроенные 
методы (такие как тренировка, прогнозирование и подбор), общие для всех классификаторов Sklearn. 
Следовательно, это исключает избыточность, поскольку не нужно писать одни и те же методы пять раз, если мы 
хотим вызвать пять различных классификаторов.
```

```python
# Некоторые полезные параметры, которые пригодятся позже
ntrain = train.shape[0]
ntest = test.shape[0]
print(ntrain)
print(ntest)
```

```python
SEED = 0 # for reproducibility
NFOLDS = 5 # set folds for out-of-fold prediction
#kf = KFold(ntrain, n_folds = NFOLDS, random_state=SEED) #что было раньше
# https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.KFold.html
kf = KFold(n_splits = NFOLDS, random_state=SEED)
```

```python
# Class to extend the Sklearn classifier
class SklearnHelper(object):
    def __init__(self, clf, seed=0, params=None):
        params['random_state'] = seed
        self.clf = clf(**params)

    def train(self, x_train, y_train):
        self.clf.fit(x_train, y_train)

    def predict(self, x):
        return self.clf.predict(x)
    
    def fit(self,x,y):
        return self.clf.fit(x,y)
    
    def feature_importances(self,x,y):
        print(self.clf.fit(x,y).feature_importances_)
    
# Класс для расширения классификатора XGboost
```

```python
Терпите меня для тех, кто уже знает это, но для людей, которые ранее не создавали классов или объектов в 
Python, позвольте мне объяснить, что делает приведенный выше код. При создании моих базовых классификаторов 
я буду использовать только модели, уже представленные в библиотеке Sklearn, и, следовательно, только расширю 
класс для этого.

def init: стандарт Python для вызова конструктора по умолчанию для класса. Это означает, что когда вы хотите 
создать объект (классификатор), вы должны указать ему параметры clf (какой классификатор sklearn вы хотите), 
seed (random seed) и params (параметры для классификаторов).

Остальная часть кода - это просто методы класса, которые просто вызывают соответствующие методы, 
уже существующие в классификаторах sklearn. По сути, мы создали класс-оболочку для расширения 
различных классификаторов Sklearn, так что это должно помочь нам сократить необходимость писать один и тот 
же код снова и снова, когда мы реализуем несколько учеников для нашего накопителя.
```

```python
Прогнозы вне фолда

Теперь, как упоминалось выше во вводном разделе, в стеке используются прогнозы базовых классификаторов в 
качестве входных данных для обучения модели второго уровня. Однако нельзя просто обучить базовые модели 
полным данным обучения, сгенерировать прогнозы для полного набора тестов и затем вывести их для обучения 
второго уровня. Это создает риск того, что ваши базовые модели предсказаний уже «увидят» набор тестов и, 
следовательно, переобучатся при подаче этих прогнозов.
```

```python
def get_oof(clf, x_train, y_train, x_test):
    oof_train = np.zeros((ntrain,))
    oof_test = np.zeros((ntest,))
    oof_test_skf = np.empty((NFOLDS, ntest))

    for i, (train_index, test_index) in enumerate(kf.split(x_train)):
        print(i)
        x_tr = x_train[train_index]
        y_tr = y_train[train_index]
        x_te = x_train[test_index]

        clf.train(x_tr, y_tr)

        oof_train[test_index] = clf.predict(x_te)
        oof_test_skf[i, :] = clf.predict(x_test)

    oof_test[:] = oof_test_skf.mean(axis=0)
    return oof_train.reshape(-1, 1), oof_test.reshape(-1, 1)
```

```python
параметры
Просто краткое изложение параметров, которые мы будем перечислять здесь для полноты,
1) n_jobs: количество ядер, используемых для тренировочного процесса. Если установлено значение -1, 
    используются все ядра.
2) n_estimators: количество деревьев классификации в вашей модели обучения (по умолчанию 10)
3) max_depth: максимальная глубина дерева, или насколько узел должен быть расширен. Остерегайтесь, если 
    установлено слишком большое число, это может привести к переобучению, так как дерево будет расти слишком 
    глубоко.
4) verbose: определяет, хотите ли вы выводить какой-либо текст в процессе обучения. Значение 0 подавляет весь 
    текст, а значение 3 выводит древовидный процесс обучения на каждой итерации.

Пожалуйста, ознакомьтесь с полным описанием на официальном сайте Sklearn. Там вы найдете целый ряд других 
полезных параметров, с которыми вы можете поиграть.
```
